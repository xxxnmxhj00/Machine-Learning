```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

# 분석에 사용할 속성을 선택

```python
ndf = rdf[['survived','pclass','sex','age','sibsp','parch','embarked']]
ndf.head()
```

```python
ndf.info()
```

```python
# pclass도 범주형 성별 찹승도시 원핫인코딩 해줘야함, 성별은 이진, 하고나서 하나의 특성을 없애주는 것도 맞겠다
```

```python
onehot_sex = pd.get_dummies(ndf['sex']).astype(int) # 처음에 TF로 나오는데 이걸 astype 해주면 된다
#ndf의 'sex' 열에 대해 원-핫 인코딩을 수행합니다. 이는 'sex' 열의 고유한 값(예: 'male', 'female')에 대해 각각의 값을 열로
# 변환하고, 해당 값이 존재하는 경우 1, 그렇지 않은 경우 0으로 표시합
onehot_sex.head()
```

```python
ndf = pd.concat([ndf, onehot_sex], axis =1)
ndf.head()
```

```python
onegot_embarked = pd.get_dummies(ndf['embarked'],prefix ='town').astype('int')
onegot_embarked.head()
```

```python
ndf = pd.concat([ndf, onegot_embarked], axis =1)
ndf.head()
```

```python
ndf.drop(['sex','embarked','male'], axis =1, inplace =True)
ndf.head()
```

# 데이터 셋 구분 - 훈련용 train data/ 검증용 test data

```python
X=ndf[list(ndf.columns)[1:]] # 독립 변수 x
y=ndf[list(ndf.columns)[0]]
```

```python
X.shape
```

```python
y.shape
```

```python
X.head()
```

```python
# 특성 정규화
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, random_state=10)
```

```python
print('train data 개수: ', X_train.shape)
print('test data 개수: ', X_test.shape)
```

```python
print('y_train data 개수: ', y_train.shape)
print('y_test data 개수: ', y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 모형 객체 생성(k=5 생성)
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)
```

```python
# 학습을 다 했으면 평가를 해야함
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve

def get_clf_eval(y_test=None , pred = None): # 여기는 우리가 만든거
    confusion = confusion_matrix(y_test , pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)

    roc_auc = roc_auc_score(y_test, pred)
    print ('오차 행렬')
    print(confusion)
    # ROC-AUC print 추가
    print('정확도 : {0: .4f}, 정밀도 : {1:.4f}, 재현율 : {2:.4f},\
          F1:{3:.4f}, AUC:{4:.4f}'.format(accuracy,precision, recall, f1, roc_auc))
```

```python
y_hat = knn.predict(X_test)
```

```python
get_clf_eval(y_test,y_hat)
```

```python
y_hat[:10]
```

```python
import numpy as np
y_test = np.array(y_test)
y_test[:10]
```

```python
from sklearn.metrics import classification_report
# 모형 성능 평가 - 평가지표 개선
knn_report = classification_report(y_test, y_hat)
print(knn_report)
```
